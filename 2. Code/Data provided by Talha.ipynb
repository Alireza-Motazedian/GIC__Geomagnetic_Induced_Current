## 4.1. SuperMAG

````python
path_SuperMAG = "/content/drive/MyDrive/5. Research/1. Summer Project/2. Data/OneDrive_1_5-22-2023/New_OTT (SuperMAG)"
````

````python
files = os.listdir(path_SuperMAG)
files_csv = [item for item in files if item[-4:].lower()==".csv"]
files_csv.sort()
files_y = [item for item in files_csv if item[-6:].lower()=="hc.csv"]
files_x = [item for item in files_csv if item[-6:].lower()!="hc.csv"]
files_y
````

````python
dfx = pd.read_csv(os.path.join(path_SuperMAG, files_x[0])) ; display("dfx:", dfx.shape, dfx.head(1)) ; print()
dfx.isna().sum()
````

````python
dfy = pd.read_csv(os.path.join(path_SuperMAG, files_y[0]), header=None) ; display("dfy:", dfy.shape, dfy.head(1)) ; print()
dfy.columns = ["y"]
dfy.isna().sum()
````

````python
dfxy = pd.concat([dfx, dfy], axis=1) ; display("dfxy:", dfxy.shape, dfxy.head(1)) ; print()
````

````python
for itemx, itemy in zip(files_x, files_y):
    dfx = pd.read_csv(os.path.join(path_SuperMAG, itemx))
    dfy = pd.read_csv(os.path.join(path_SuperMAG, itemy), header=None)
    dfy.columns = ["y"]
    dfxy = pd.concat([dfx, dfy], axis=1) ; display("dfxy:", dfxy.shape, dfxy.head(1)) ; print()
    display(dfxy.isna().sum()) ; print()
````

- many nones
- 525600 =
Common year (365 days):
There are 24 hours in a day and 60 minutes in an hour. So, to find the number of minutes in a day: 24 hours x 60 minutes = 1,440 minutes.
Now, multiply the number of minutes in a day by the number of days in a common year: 1,440 minutes x 365 days = 525,600 minutes.

````python
list0 = []
for itemx, itemy in zip(files_x, files_y):
    dfx = pd.read_csv(os.path.join(path_SuperMAG, itemx))
    dfy = pd.read_csv(os.path.join(path_SuperMAG, itemy), header=None)
    dfy.columns = ["y"]
    dfxy = pd.concat([dfx, dfy], axis=1) #; display("dfxy:", dfxy.shape, dfxy.head(1)) ; print()
    #display(dfxy.isna().sum()) ; print()
    list0.append(dfxy)

df_SuperMAG = pd.concat(list0) ; display("df_SuperMAG:", df_SuperMAG.shape, df_SuperMAG.head(1)) ; print()
````
